```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
neft = pd.read_excel("Neft May 2016.xlsx")
```

```python
neft = neft.drop('Unnamed: 0', axis =1 )
```

```python
neft.columns = ['NATIONAL ELECTRONIC FUND TRANSFER (NEFT) - MAY 2016','Bank', 'Inward No. of Transations',
       'Inward Amount', 'Outward No. of Transactions', 'Outward Amount']
```

```python
neft = neft[2:174]
```

```python
neft['Total Transactions'] = neft['Inward No. of Transations']+neft['Outward No. of Transactions']
```

```python
neft
```

```python
neft.sort_values("Inward No. of Transations" , inplace=True)
```

```python
neft_2016 = pd.DataFrame({'Credit transactions':neft['Inward No. of Transations'].tolist(),'Debit transactions':neft['Outward No. of Transactions'].tolist()},index=neft['Bank'])
```

```python
neft_2016
```

```python
neft_2016.plot(kind='bar',figsize=(20,10))
```

```python
#subset the number of banks to more popular(highly used) banks
neft_2016.query('Bank == "STATE BANK OF INDIA" or Bank == "HDFC BANK" or Bank == "ICICI BANK LTD" or Bank == "AXIS BANK" or Bank == "KOTAK MAHINDRA BANK LTD"').plot(kind='bar')
```

```python
neft_2017 = pd.read_excel("Neft May 2017.xlsx")
```

```python
#neft_2017
neft_2017 = neft_2017.drop('Unnamed: 0', axis =1 )
neft_2017.columns = ['NATIONAL ELECTRONIC FUND TRANSFER (NEFT) - MAY 2017','Bank', 'Inward No. of Transations',
       'Inward Amount', 'Outward No. of Transactions', 'Outward Amount']
neft_2017=neft_2017[2:188]
```

```python
neft_2018 = pd.read_excel("Neft May 2018.xlsx")
```

```python
#neft_2018
neft_2018 = neft_2018.drop('Unnamed: 0', axis =1 )
neft_2018.columns = ['NATIONAL ELECTRONIC FUND TRANSFER (NEFT) - MAY 2017','Bank', 'Outward No. of Transations',
       'Outward Amount', 'Inward No. of Transactions', 'Inward Amount']
neft_2018=neft_2018[2:197]
```

```python
neft_2018['Total Transactions'] = neft_2018['Inward No. of Transactions']+neft_2018['Outward No. of Transations']
```

```python
neft_2018
```

```python
#subset the number of banks to more popular(highly used) banks
neft2016May = neft.query('Bank == "STATE BANK OF INDIA" or Bank == "HDFC BANK" or Bank == "ICICI BANK LTD" or Bank == "AXIS BANK" or Bank == "KOTAK MAHINDRA BANK LTD"')
neft2017May = neft_2017.query('Bank == "STATE BANK OF INDIA" or Bank == "HDFC BANK" or Bank == "ICICI BANK LTD" or Bank == "AXIS BANK" or Bank == "KOTAK MAHINDRA BANK LTD"')
neft2018May = neft_2018.query('Bank == "STATE BANK OF INDIA" or Bank == "HDFC BANK" or Bank == "ICICI BANK LTD" or Bank == "AXIS BANK" or Bank == "KOTAK MAHINDRA BANK LTD"')

neftMay = pd.DataFrame({'2016':neft2016May['Total Transactions'].tolist(),'2017':neft2017May['Total Transactions'].tolist(),'2018':neft2018May['Total Transactions'].tolist()},index=neft2017May['Bank'])
neftMay.plot(kind = 'bar',figsize=(10,7))
```
